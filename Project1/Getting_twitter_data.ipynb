```
#Estoy indentificandome con mi auth
twitter_api = tw.Twitter(auth = auth)
```

```
#Ahora voy a buscar tweets relacionado con el coronavirus. Pero en Lima y en Mexico DF. 
#(ID's sacados de yweather library)
woeid_lima = 418440
woeid_mexicodf = 116545
```
